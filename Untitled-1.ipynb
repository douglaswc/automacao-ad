```python
import os

# Define o diretório onde o arquivo está localizado
diretorio = r'C:\Criação de usuário'

# Lista todos os arquivos no diretório
arquivos_no_diretorio = os.listdir(diretorio)

# Verifica se há arquivos de texto (.txt) no diretório
arquivos_txt = [arquivo for arquivo in arquivos_no_diretorio if arquivo.endswith('.txt')]

# Verifica se há pelo menos um arquivo de texto no diretório
if arquivos_txt:
    # Assume que o primeiro arquivo de texto encontrado é o arquivo desejado
    nome_arquivo = arquivos_txt[0]

    # Caminho completo para o arquivo
    caminho_arquivo = os.path.join(diretorio, nome_arquivo)

    # Dicionário para armazenar as variáveis
    variaveis = {}

    # Abre o arquivo em modo de leitura
    with open(caminho_arquivo, 'r') as arquivo:
        # Lê todas as linhas do arquivo
        linhas = arquivo.readlines()

        # Itera sobre cada linha do arquivo
        for idx, linha in enumerate(linhas):
            # Verifica se a linha contém "Login:"
            if "Login:" in linha:
                # Captura o texto após "Login:"
                login = linha.split("Login:", 1)[-1].strip()
                
                # Substitui espaços em branco por underscores para criar o nome da variável
                nome_variavel = login.replace(" ", "_")
                
                # Se o nome da variável já existir, adiciona um número ao final
                if nome_variavel in variaveis:
                    nome_variavel += f"_{idx}"
                
                # Define a variável com o nome correspondente
                variaveis[nome_variavel] = login

    # Mostra as variáveis criadas
    print("Variáveis criadas:")
    for nome_variavel, valor in variaveis.items():
        globals()[nome_variavel] = valor
        print(f"{nome_variavel} = {valor}")

else:
    print("Nenhum arquivo de texto (.txt) encontrado no diretório especificado.")
```

```python
import random
import string

# Define uma função para gerar uma senha aleatória
def gerar_senha():
    # Define os caracteres especiais permitidos
    caracteres_especiais = '!@#$%^&*()-_=+'
    # Gera uma senha com 8 caracteres
    tamanho = 8
    # Garante que haja pelo menos um caractere especial na senha
    senha = random.choice(caracteres_especiais)
    # Adiciona um número à senha
    senha += random.choice(string.digits)
    # Adiciona uma letra minúscula à senha
    senha += random.choice(string.ascii_lowercase)
    # Gera o restante da senha com letras maiúsculas, números e caracteres especiais
    for _ in range(tamanho - 3):
        senha += random.choice(string.ascii_uppercase + string.digits)
    # Embaralha a senha para torná-la mais segura
    senha_embaralhada = ''.join(random.sample(senha, len(senha)))
    return senha_embaralhada

# Exemplo de uso da função para gerar uma senha
nova_senha = gerar_senha()
print("Nova senha gerada:", nova_senha)
```
